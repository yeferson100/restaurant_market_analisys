## REVIEWS DE GOOGLE

- Tenemos un data con 51 carpetas con los reviews de los diferentes estados de US.
- Cada carpeta contiene los reviews de un estado.

##### - Como cada estado contiene varios archivos debemos generar un solo archivo con todos los reviews por estado

```python
import pandas as pd
import os
import json
import glob
```

```python


# Ruta de la carpeta principal que contiene las carpetas por estado con archivos .json
ruta_principal = 'C:/Users/Yeferson/Documents/Henry/ProyectoFinal/reviews-estados/'

# Obtener la lista de carpetas que contienen archivos .json
carpetas = glob.glob(os.path.join(ruta_principal, 'review-*'))

for carpeta in carpetas:
    # Lista para almacenar los datos de los archivos .json de cada subcarpeta
    datos_json = []

    for archivo in os.listdir(carpeta):
        if archivo.endswith('.json'):
            ruta_archivo = os.path.join(carpeta, archivo)
            with open(ruta_archivo, 'r') as file:
                for linea in file:
                    try:
                        datos_archivo = json.loads(linea)
                        datos_json.append(datos_archivo)
                    except json.JSONDecodeError:
                        print(f"Error al decodificar JSON en el archivo: {ruta_archivo}")

    # Nombre del archivo combinado dentro de la subcarpeta
    archivo_combinado = os.path.join(carpeta, 'combined.json')

    # Guardar los datos combinados en el archivo .json dentro de la subcarpeta
    with open(archivo_combinado, 'w') as file:
        json.dump(datos_json, file, indent=4)

    print(f"Se han combinado todos los archivos .json en '{archivo_combinado}'.")
```

* Leemos como DATAFRAME cada archivo y asignamos las abreviaciones del estado al que pertenecen.
* Tenemos valores Time en ms y pasamos a valores tipo Fecha.
* Almacenamos en archivos parquet para disminuir su tama침o.

```python
ruta_archivo ='C:/Users/Yeferson/Documents/Henry/ProyectoFinal/reviews-estados/review-Alabama/combined.json'

data = pd.read_json(ruta_archivo)

data['state']='VA' # Para facilitar su clasificaci칩n o identificaci칩n

data['time'] = pd.to_datetime(data['time'], unit='ms') #de ms a Date

#Almacenamos el data en formato parquet, para disminuir el tama침o del archivo
data.to_parquet('alabama.parquet',compression='brotli')
```
