```python
#Los archivos van desde 1.json hasta 11.json
#Debido al tama√±o de los archivos es necesario aplicar una transformacion por cada archivo
import pandas as pd
meta_google1=pd.read_json('/content/drive/MyDrive/Google Maps/metadata-sitios/11.json', lines=True)
#Cambiamos el archivo final para cada archivo en nuestra carpeta /metadata_sitios/
```

```python
#Eliminamos  los valores duplicados por ID de usuario
meta_google1.loc[meta_google1.duplicated(subset=['gmap_id'])]
meta_google1 = meta_google1.loc[~meta_google1.duplicated(subset=['gmap_id'])] .reset_index(drop=True).copy()
meta_google1.info()
```

```python
#Creamos un campo para asignar un estado
meta_google1['estado']=None
```

```python
#Agregamos las abreviaciones que corresponden a cada estado

meta_google1.loc[meta_google1['address'].str[-8:-6]=='MD', 'estado']='MD'
meta_google1.loc[meta_google1['address'].str[-8:-6]=='GA', 'estado']='GA'
meta_google1.loc[meta_google1['address'].str[-8:-6]=='VA', 'estado']='VA'
meta_google1.loc[meta_google1['address'].str[-8:-6]=='NC', 'estado']='NC'
meta_google1.loc[meta_google1['address'].str[-8:-6]=='SC', 'estado']='SC'
meta_google1.loc[meta_google1['address'].str[-8:-6]=='FL', 'estado']='FL'
```

```python
#Seleccionamos los comercios que correspondan a los estados seleccionados GA, MD, VA, NC, SC, FL

meta_state=meta_google1[(meta_google1['address'].str[-8:-6]=='GA')|(meta_google1['address'].str[-8:-6]=='MD')|
                (meta_google1['address'].str[-8:-6]=='VA')|(meta_google1['address'].str[-8:-6]=='NC')|
                (meta_google1['address'].str[-8:-6]=='SC')|(meta_google1['address'].str[-8:-6]=='FL')]
meta_state
```

```python
meta_state.to_csv('meta11.csv', index=False)
```
